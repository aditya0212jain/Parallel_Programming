```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
# sequential cluster points after k means
d1 = pd.read_csv('seq_a.txt', sep=" " , usecols = [0,1,2,3],header = None)
d1.columns =['x','y','z','cluster']
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
k = 10
colors = {1:"red",2:"blue",3:"violet",4:"black",5:"green"}
for i in range(k):
    d2 = d1.loc[d1['cluster'] == i]
    x = list(d2['x'])
    y = list(d2['y'])
    z = list(d2['z'])
    ax.scatter(x,y,z)
    
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.title("Sequential 1000k points")

plt.show()
```

```python
d3 = pd.read_csv('eplot3.txt',sep=" ",header=None)
d4 = pd.read_csv('eplot4.txt',sep=" ",header=None)
num = [2,3,5,6,10,16,32]
plt.plot(num, list(d3[0]),label="pThread")
plt.plot(num, list(d4[0]),label="openMP")
plt.legend()
plt.xlabel('Number of Processors')
plt.ylabel('Efficiency')
plt.title('Efficiency vs Number of Processors')
plt.show()
```
